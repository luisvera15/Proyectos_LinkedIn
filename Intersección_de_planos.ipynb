```python
# importamos las librerias necesarias
import matplotlib.pyplot as plt
import mplstereonet
%matplotlib inline

# creamos los planos
strike1, dip1 = 215, 30
strike2, dip2 = 120, 40

fig, ax = mplstereonet.subplots()
ax.plane(strike1, dip1,label='Plano1')
ax.plane(strike2, dip2,label='Plano2')
# encontramos la interseccin de ambos planos y graficamos
plunge, bearing = mplstereonet.plane_intersection(strike1, dip1, strike2, dip2)
ax.line(plunge, bearing, 'ko', markersize=5, 
        label='Interseccin %02d/%03d' % (plunge, bearing))
# aadimos grid y la leyenda
ax.grid()
ax.legend(loc='upper right')
plt.show()
```
